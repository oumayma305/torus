```python
import math
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sympy.plotting import plot3d_parametric_surface
from sympy import sin, cos
from sympy.abc import k, h

def f(x,y):
    return x**2 + y**2 + x*y
X = np.arange(-1, 1, 0.01)
Y = np.arange(-1, 1, 0.01)
X, Y = np.meshgrid(X, Y)
Z = f(X, Y)
plt.title('flat torus')
plt.contour(X, Y, Z, [0.1,0.4,0.5])
plt.show()


n = 100
theta = np.linspace(0, 4.*np.pi, n)
phi = np.linspace(0, 4.*np.pi, n)
theta, phi = np.meshgrid(theta, phi)
a, b = 2, 1
x = (a + b*np.cos(theta)) * np.cos(phi)
y = (a + b*np.cos(theta)) * np.sin(phi)
z = b * np.sin(theta)
fig = plt.figure()
ax1 = fig.add_subplot(121, projection='3d')
ax1.set_zlim(-10,10)
ax1.plot_surface(x, y, z, rstride=5, cstride=5, color='M', edgecolors='w')
ax1.view_init(40, 40)
ax2 = fig.add_subplot(122, projection='3d')
ax2.set_zlim(-11,11)
ax2.plot_surface(x, y, z, rstride=5, cstride=5, color='R', edgecolors='w')
ax2.view_init(0, 0)
ax2.set_xticks([])
plt.show()


X = cos(k)*(3.5+2.5*cos(h))
Y = sin(k)*(3.5+2.5*cos(h))
Z = sin(h)
plot3d_parametric_surface(X, Y, Z, (k, -0.5, 6), (h, -0.5, 6))


```
